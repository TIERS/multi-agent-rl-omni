```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
custom_params = {
    "lines.linewidth": 2.0,
    # "xtick.bottom": True,
    # "ytick.left": True,
    # "xtick.top": True,
    # "ytick.right": True,
 }
sns.set(font_scale=1.5, rc=custom_params)
```

```python
# experiment
e = 3

# p1
baseline = pd.read_csv("data/day2/e{}_p1_baseline.csv".format(e))
m1 = pd.read_csv("data/day2/e{}_p1_m1.csv".format(e))
m2 = pd.read_csv("data/day2/e{}_p1_m2.csv".format(e))

# p2
baseline2 = pd.read_csv("data/day2/e{}_p2_baseline.csv".format(e))
m12 = pd.read_csv("data/day2/e{}_p2_m1.csv".format(e))
m22 = pd.read_csv("data/day2/e{}_p2_m2.csv".format(e))

# p3
baseline3 = pd.read_csv("data/day2/e{}_p3_baseline.csv".format(e))
m13 = pd.read_csv("data/day2/e{}_p3_m1.csv".format(e))
m23 = pd.read_csv("data/day2/e{}_p3_m2.csv".format(e))
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.plot(baseline["elapsed_time"], baseline["distance_from_target"], label="baseline")
plt.plot(m1["elapsed_time"], m1["distance_from_target"], label="m1")
plt.plot(m2["elapsed_time"], m2["distance_from_target"], label="m2")
plt.ylabel("Distance from target (m)")
plt.xlabel("Time (s)")
plt.title("Point 1")

plt.subplot(132)
plt.plot(baseline2["elapsed_time"], baseline2["distance_from_target"], label="baseline")
plt.plot(m12["elapsed_time"], m12["distance_from_target"], label="m1")
plt.plot(m22["elapsed_time"], m22["distance_from_target"], label="m2")
plt.xlabel("Time (s)")
plt.title("Point 2")
#plt.title("Distance from target over time")

plt.subplot(133)
plt.plot(baseline3["elapsed_time"], baseline3["distance_from_target"], label="baseline")
plt.plot(m13["elapsed_time"], m13["distance_from_target"], label="m1")
plt.plot(m23["elapsed_time"], m23["distance_from_target"], label="m2")
plt.xlabel("Time (s)")
plt.title("Point 3")

# add labels


# add title


plt.legend()
#plt.show()

plt.savefig("experiment{}.pdf".format(e), bbox_inches='tight')

#import tikzplotlib

#tikzplotlib.save("mytikz.tex")

```
